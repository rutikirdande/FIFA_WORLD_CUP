```python
from scipy import stats

# Hypothesis test: Is there a significant difference in 'Attendance' between two years?
year_1_attendance = df1[df1['Year'] == 2018]['Attendance']
year_2_attendance = df1[df1['Year'] == 2014]['Attendance']

t_stat, p_value = stats.ttest_ind(year_1_attendance, year_2_attendance)

if p_value < 0.05:
    print("There is a significant difference in attendance between the two years.")
else:
    print("There is no significant difference in attendance between the two years.")
```

```python
# Calculate the correlation between 'Attendance' and 'Goals Scored'
correlation = df3['Attendance'].corr(df3['GoalsScored'])
print(f"Correlation between Attendance and Goals Scored: {correlation}")
```

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df1['Stage'], df3['Winner'])

# Perform the chi-square test
chi2, p, _, _ = chi2_contingency(contingency_table)

# Set the significance level
alpha = 0.05

if p < alpha:
    print(f"There is a significant association between 'Stage' and 'Winner' (p-value: {p})")
else:
    print(f"No significant association between 'Stage' and 'Winner' (p-value: {p})")
```

```python
# Calculate the Pearson correlation coefficient
correlation = df1['Home Team Goals'].corr(df1['Attendance'])

# Interpret the correlation coefficient
if correlation > 0:
    interpretation = "positive"
elif correlation < 0:
    interpretation = "negative"
else:
    interpretation = "no"

# Output the results
print(f"Pearson Correlation Coefficient between Home Team Goals and Attendance: {correlation:.2f}")
print(f"There is a {interpretation} correlation between Home Team Goals and Attendance.")
```

```python
from scipy.stats import ttest_ind

# Separate attendance data for Final and Semi-finals matches
final_attendance = df1[df1['Stage'] == 'Final']['Attendance']
semifinals_attendance = df1[df1['Stage'] == 'Semi-finals']['Attendance']

# Perform a two-sample T-Test
t_stat, p_value = ttest_ind(final_attendance, semifinals_attendance)

# Set significance level
alpha = 0.05

# Interpret the results
if p_value < alpha:
    interpretation = "reject the null hypothesis"
else:
    interpretation = "fail to reject the null hypothesis"

# Output the results
print(f"T-Test Statistic: {t_stat:.2f}")
print(f"P-Value: {p_value:.4f}")
print(f"At a significance level of {alpha}, we {interpretation}.")
```

```python


# Create an empty DataFrame to store position counts
position_counts = pd.DataFrame(columns=['Team', 'Winner', 'Runners-Up', 'Third', 'Fourth'])

# Iterate through each row in the dataset and extract positions for each team
for _, row in df3.iterrows():
    positions = ['Winner', 'Runners-Up', 'Third', 'Fourth']
    for position in positions:
        team = row[position]
        position_counts = position_counts.append({'Team': team, position: 1}, ignore_index=True)

# Fill missing values with 0
position_counts = position_counts.fillna(0)

# Group and sum the counts by team
position_counts = position_counts.groupby('Team').sum()

# Sort the data by the total number of positions
position_counts['Total'] = position_counts.sum(axis=1)
position_counts = position_counts.sort_values('Total', ascending=False)

# Create a stacked bar chart to visualize positions for each team in different seasons
teams = position_counts.index
x = range(len(teams))
bar_width = 0.6

plt.figure(figsize=(12, 6))

positions = ['Winner', 'Runners-Up', 'Third', 'Fourth']
bottom = [0] * len(teams)

for position in positions:
    plt.bar(x, position_counts[position], bar_width, label=position, alpha=0.7, bottom=bottom)
    bottom = [bottom[i] + position_counts[position][i] for i in range(len(teams))]

plt.xlabel('Team')
plt.ylabel('Count')
plt.title('Team Positions in Different Seasons')
plt.xticks(x, teams, rotation=90)
plt.legend()
plt.show()
```
